```python
import os
import urllib.request

#Importação para barra de progresso:
from tqdm import tqdm
```

```python
# Caminho para o arquivo de texto com os links
caminho_arquivo = input('Caminho onde está o seu txt com os links dos arquivos HDF:\n')

# Caminho para a pasta onde os arquivos serão salvos
caminho_pasta = input('Digite o caminho onde os arquivos HDF ficaram salvos:\n')

# Cria a pasta de downloads se ela não existir
if not os.path.exists(caminho_pasta):
    os.makedirs(caminho_pasta)

# Lê os links do arquivo de texto
with open(caminho_arquivo, 'r') as arquivo:
    links = arquivo.read().splitlines()

# Itera sobre os links e faz o download dos arquivos
for link in tqdm(links):
    nome_arquivo = link.split('/')[-1]
    caminho_arquivo = os.path.join(caminho_pasta, nome_arquivo)
    urllib.request.urlretrieve(link, caminho_arquivo)
```
